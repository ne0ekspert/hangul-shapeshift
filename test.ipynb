```python
before = "것"
after = "닭"
b = list(j2hcj(h2j(before)))
a = list(j2hcj(h2j(after)))

while len(b) != 3:
    b.append(' ')
while len(a) != 3:
    a.append(' ')

print(f"cho: {nx.shortest_path(G_cho, b[0], a[0])}")
print(f"jung: {nx.shortest_path(G_jung, b[1], a[1])}")
print(f"jong: {nx.shortest_path(G_jong, b[2], a[2])}")
```
